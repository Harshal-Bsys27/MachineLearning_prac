```python
import pandas as pd
```

```python
df = pd.read_csv("student_performance.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
# Fill numeric column with mean
df['total_score'].fillna(df['total_score'].mean(), inplace=True)

# Fill categorical column with mode
df['grade'].fillna(df['grade'].mode()[0], inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['grade'] = df['grade'].map({'A':4, 'B':3, 'C':2, 'D':1, 'F':0})
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.to_csv("cleaned_student_data.csv", index=False)
```

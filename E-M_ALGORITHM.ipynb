```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
```

```python
# Step 1: Load dataset
df = pd.read_csv("iris.csv")
```

```python
df.head()
```

```python
# Step 2: Select features
X = df[['sepal_length', 'sepal_width']]
```

```python
# Step 3: Handle missing values (important for EM)
X = X.dropna()
```

```python
# Step 4: Apply EM Algorithm using Gaussian Mixture Model
gmm = GaussianMixture(n_components=2, random_state=42)
gmm.fit(X)
```

```python
# Step 5: Predict cluster labels
labels = gmm.predict(X)
```

```python
# Step 6: Visualize clusters
plt.figure()
plt.scatter(X['sepal_length'], X['sepal_width'], c=labels)
plt.xlabel("sepal_length")
plt.ylabel("sepal_width")
plt.title("Expectation-Maximization (EM) Clustering")
plt.show()
```
